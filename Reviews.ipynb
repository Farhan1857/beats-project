```python
# Importing pandas
import pandas as pd
import numpy as np
```

```python
#Upload the CSV file and read it into a DataFrame
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("/content/beats_wireless_speakers_reviews_1000.csv")
df = df.dropna(subset=["content"])
```

```python
# Inspect the first few rows of the DataFrame
print("First five rows of the dataset:")
print(df.head())

# Get a summary of the DataFrame
print("\nDataFrame summary:")
print(df.info())
```

```python
import pandas as pd

# 1) Detect missing values
print(df.isnull().sum())

# 2) Handle missing values
# Only 'content' has 1 missing value, so just drop that row
df = df.dropna(subset=["content"])

print("Shape after dropping missing content:", df.shape)

# Double-check missing values again
print(df.isnull().sum())
```

```python
import pandas as pd

# 1) Identify duplicate rows (entire row duplicates)
duplicates = df.duplicated()
print("Duplicate rows (True = duplicate):\n", duplicates)

print("\nNumber of duplicate rows:", duplicates.sum())

# 2) Remove duplicates
df = df.drop_duplicates()

print("\nShape after removing duplicates:", df.shape)
```

```python
import pandas as pd

# Correcting data types

# rating should be numeric (float or int)
df["rating"] = df["rating"].astype(float)

# helpful_count should be integer
df["helpful_count"] = df["helpful_count"].astype(int)

# is_verified should be boolean
df["is_verified"] = df["is_verified"].astype(bool)

# timestamps as strings
df["review_date"] = df["review_date"].astype(str)
df["timestamp"] = df["timestamp"].astype(str)

print(df.dtypes)
```

### **Starting Numpy Steps**


```python
df.head()
```

```python
# Convert 'rating' column to a NumPy array
ratings = df["rating"].values

# Replace missing values with mean using NumPy
mean_rating = np.nanmean(ratings)
ratings = np.where(np.isnan(ratings), mean_rating, ratings)

# Assign back to DataFrame
df["rating"] = ratings
```

```python
# Keep only ratings between 1 and 5
valid_ratings = ratings[(ratings >= 1) & (ratings <= 5)]
df = df[df["rating"].isin(valid_ratings)]

# Update ratings array after filtering
ratings = df["rating"].values
```

```python
# Calculate review lengths
review_lengths = np.array([len(str(review).split()) for review in df["content"]])
```

```python
min_rating = np.min(ratings)
max_rating = np.max(ratings)

normalized_ratings = (ratings - min_rating) / (max_rating - min_rating)

df["normalized_rating"] = normalized_ratings
```

```python
df["rating"] = df["rating"].astype(np.float64)
```

```python
df["review_length"] = review_lengths
df.to_csv("cleaned_reviews_full.csv", index=False)
```
