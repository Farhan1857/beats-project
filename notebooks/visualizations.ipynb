```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/cleaned_reviews_full.csv")
df.head()
```

```python
sns.histplot(df["rating"], bins=5, kde=True)
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x="brand", data=df, order=df["brand"].value_counts().index)
plt.title("Number of Reviews by Brand")
plt.xlabel("Brand")
plt.ylabel("Review Count")
plt.xticks(rotation=45)
plt.show()
```

```python
avg_brand_rating = df.groupby("brand")["rating"].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=avg_brand_rating.index, y=avg_brand_rating.values)
plt.title("Average Rating by Brand")
plt.xlabel("Brand")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.ylim(0, 5)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x="review_length", y="rating", hue="brand", data=df)
plt.title("Review Length vs Rating (by Brand)")
plt.xlabel("Review Length (words)")
plt.ylabel("Rating")
plt.ylim(0, 5.5)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x="brand", y="rating", data=df)
plt.title("Rating Distribution by Brand (Box Plot)")
plt.xlabel("Brand")
plt.ylabel("Rating")
plt.xticks(rotation=45)
plt.ylim(0, 5)
plt.show()
```
