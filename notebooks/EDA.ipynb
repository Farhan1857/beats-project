```python
import pandas as pd

df = pd.read_csv("/content/cleaned_reviews_full.csv")
```

```python
print(df.head())
print(df.tail())
```

```python
df.describe()
```

```python
# Showing the descriptive statistics for the rating column
print("Mean:", df["rating"].mean())
print("Median:", df["rating"].median())
print("Mode:", df["rating"].mode()[0])
print("Variance:", df["rating"].var())
print("Standard Deviation:", df["rating"].std())
```

```python
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
if df["rating"].isnull().sum() > 0:
    df["rating"].fillna(df["rating"].mean(), inplace=True)
```

```python
Q1 = df["rating"].quantile(0.25)
Q3 = df["rating"].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df["rating"] < (Q1 - 1.5 * IQR)) | (df["rating"] > (Q3 + 1.5 * IQR))]
print("Outliers in rating:\n", outliers)
print("\nNumber of rating outliers:", outliers.shape[0])
```

```python
# Calculating outliers in review_length
if "review_length" in df.columns:
    Q1 = df["review_length"].quantile(0.25)
    Q3 = df["review_length"].quantile(0.75)
    IQR = Q3 - Q1

    outliers_len = df[(df["review_length"] < (Q1 - 1.5 * IQR)) | (df["review_length"] > (Q3 + 1.5 * IQR))]
    print("\nOutliers in review_length:\n", outliers_len.head())
    print("\nNumber of review_length outliers:", outliers_len.shape[0])
```
